## Data Cleaning

### Question-1:

```python
null_citric_acid = wine['citric acid'].isnull().sum()
print(f"Number of null values in citric acid column is {null_citric_acid}")
```

Replacing null values of citric acid column with the median value of that column:

```python
wine['citric acid'].fillna(wine['citric acid'].median(), inplace=True)
```

Checking if the values are filled in:

```python
null_citric_acid = wine['citric acid'].isnull().sum()
print(f"Number of null values in citric acid column is {null_citric_acid}")
```

### Question-2:

Checking how many null values are present in residual sugar column:

```python
null_residual_sugar = wine['residual sugar'].isnull().sum()
print(f"Number of null values in residual sugar column is {null_residual_sugar}")
```

Removing all rows where the residual sugar column has null values:

```python
wine.dropna(subset='residual sugar', inplace=True)
```

Checking if the dataset is modified:

```python
num_of_rows = len(wine)
print(f"Number of rows in wine dataset is {num_of_rows}")
```

111 rows are removed.

### Question-3:

Normalization formula: normalized(x) = x-min(x) / max(x)-min(x)

```python
wine['fixed acidity'] = (wine['fixed acidity']-wine['fixed acidity'].min()) / (wine['fixed acidity'].max()-wine['fixed acidity'].min())
```

Checking the dataset:

```python
wine.describe()
```

Fixed acidity column is successfully normalized between 0 and 1.

### Question-4:

Checking no. of Duplicate rows:

```python
dup = wine.duplicated().sum()
print(f"Number of duplicates in wine dataset is {dup}")
```

Removing Duplicates:

```python
wine.drop_duplicates(inplace=True)
```

Checking if duplicate rows are removed:

```python
num_of_original = len(wine)
print(f"Number of rows in wine dataset is {num_of_original}")
```

136 rows Removed.
